| famsup | 가족 학습 지원 여부 | 가족의 학습 지원 여부 (yes: 있음, no: 없음) |
| paid | 유료 수업 수강 여부 | 과목별 유료 수업 수강 여부 (yes: 있음, no: 없음) |
| activities | 과외 활동 참여 여부 | 교과 외 활동 참여 여부 (yes: 있음, no: 없음) |
| nursery | 유치원 다님 여부 | 유치원 교육 경험 여부 (yes: 있음, no: 없음) |
| higher | 고등교육 희망 여부 | 고등교육 진학 의향 (yes: 있음, no: 없음) |
| internet | 인터넷 접근 가능 여부 | 집에서 인터넷 사용 가능 여부 (yes: 가능, no: 불가능) |
| romantic | 연인 관계 여부 | 연인 관계 상태 (yes: 있음, no: 없음) |
| famrel | 가족 관계 품질 | 가족 관계의 질 (1: 매우 나쁨 ~ 5: 매우 좋음) |
| freetime | 여가 시간 | 방과 후 여가 시간 (1: 매우 적음 ~ 5: 매우 많음) |
| goout | 외출 빈도 | 친구들과의 외출 빈도 (1: 매우 적음 ~ 5: 매우 자주) |
| Dalc | 평일 음주량 | 평일 알코올 소비량 (1: 매우 적음 ~ 5: 매우 많음) |
| Walc | 주말 음주량 | 주말 알코올 소비량 (1: 매우 적음 ~ 5: 매우 많음) |
| health | 건강 상태 | 현재 건강 상태 (1: 매우 나쁨 ~ 5: 매우 좋음) |
| absences | 결석 횟수 | 학교 결석 횟수 (0-93회) |

**데이터셋 특징**
이 데이터셋은 인도 교육 시스템을 기반으로 한 것으로 보입니다 (카스트 시스템, 10학년/12학년 구분 등).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. 데이터 준비
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
performance = fetch_ucirepo(id=320)
higher_students_performance = fetch_ucirepo(id=856) 
student_examination = fetch_ucirepo(id=582)  

performance_features = performance.data.features
performance_target = performance.data.target


higher_students_performance_features = higher_students_performance.data.features
higher_students_performance_target = higher_students_performance.data.target

student_examination_features = student_examination.data.features
student_examination_target = student_examination.data.target

# 2. 데이터프레임 생성
student_data_df = pd.DataFrame(data=performance_features, columns=performance.data.feature_names)
higher_students_df = pd.DataFrame(data=higher_students_performance_features, columns=higher_students_performance.data.feature_names)
student_education_df = pd.DataFrame(data=student_examination_features, columns=student_examination.data.feature_names)


# CSV 파일로 저장
student_data_df.to_csv('student_data.csv', index=False)
higher_students_df.to_csv('higher_students.csv', index=False)
student_education_df.to_csv('student_education.csv', index=False)

print(student_education_df.columns)
```
